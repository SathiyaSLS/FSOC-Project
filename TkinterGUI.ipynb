```python
from tkinter import *
from tkinter import ttk, filedialog
from tkinter.filedialog import askopenfile, askdirectory, askopenfilename
from tkinter.scrolledtext import ScrolledText
```

```python
def browseDir():
    fileDir.delete(0, END)
    tfFileDir = filedialog.askopenfilename()
    fileDir.insert(0, tfFileDir)


def transferFn():
    if (fileDir.get() == "") or (fileDir.get() == " "):
        errorLog("Please Enter File Directory")
    else:
        data = "File: " + fileDir.get() + "\n"
        labelLog.delete("end-2l", "end-1l")
        labelLog.insert(END, data)
        print(fileDir.get())
def errorLog(data):
    labelLog.insert(END, "\tERROR: "+data+"\n")
    labelLog.tag_add("start", "end-2l", "end-1l")
    labelLog.tag_config("start", foreground="red", font=("Myriad Pro Light", 10, 'bold'))

def appendLog(data):
    oldText = labelLog.cget("text")
    newText = oldText + data
    labelLog.insert(END, data)
```

```python
root = Tk()

bgcolor = '#E4DCCF'
wnW = root.winfo_screenwidth()
wnH = root.winfo_screenheight()
width = 600
height = 400

root.title("Optic File Transfer")
root.geometry('%dx%d+%d+%d' % (width, height, ((wnW / 2) - (width / 2)),
                               ((wnH / 2) - (height / 2))))
root.configure(bg=bgcolor)

favicon = PhotoImage(file='favicon.png')
# root.iconphoto(False, favicon)

window = Frame(root, bg=bgcolor)
window.pack(expand=1, fill=BOTH, side=BOTTOM, padx=20, pady=20)

frame1 = Frame(window, bg=bgcolor)
frame1.pack(expand=1, side=TOP, anchor='n')
frame12 = Frame(frame1, bg=bgcolor)
frame12.pack(expand=1, fill=X, side=BOTTOM, anchor='center')
frame11 = Frame(frame1, bg='black', bd=2)
frame11.pack(expand=1, side=BOTTOM, anchor='center')

labelSelectFile = Label(frame1,
                        text='Select The File :',
                        font=("Myriad Pro Light", 10),
                        bg=bgcolor)

fileDir = Entry(
    frame11,
    font=("Myriad Pro Bold", 10),
    width=40,
    bd=0,
    #bg='#f1f3f6'
)
browseBtn = Button(frame11,
                   font=("Myriad Pro Light", 10),
                   text="Browse File",
                   bd=0,
                   bg='#e0b4a4',
                   command=browseDir)
transferBtn = Button(frame12,
                     font=("Myriad Pro", 10, 'bold'),
                     text="Transfer >>",
                     relief='groove',
                     bg='#22404a',
                     fg='#fff2f2',
                     command=transferFn)
# labelLog = Text(
#     frame12,
#     relief='groove',
#     font=("Myriad Pro Light", 8),
#     bg='#fefaf1',
#     width=64,
#     height=10,)

labelLog = ScrolledText(frame12,
                        relief='groove',
                        width=64,
                        height=10,
                        font=("Myriad Pro Light", 8),
                        bg='#fefaf1')

labelLog.insert(END, "LOG:\nBrowse the File...\n")

labelSelectFile.pack(side=TOP, anchor='w')
fileDir.pack(side=LEFT, ipadx=20, ipady=7)
browseBtn.pack(side=LEFT, ipadx=2, ipady=4)
transferBtn.pack(side=TOP, padx=10, pady=10, ipadx=11, ipady=4, anchor='e')
labelLog.pack(side=BOTTOM, pady=10, anchor='n')

mainloop()
```
