```python
def bin2Data(bin_data):
    data = ''
    bit = ''
    for i in range(len(bin_data)):
        bit += bin_data[i]
        if ((i+1)%8 == 0):
            data += chr(int(bit, 2))
            # data = "".join(chr(int(bit, 2)))
            bit = ''
    return data

def writeFile_2(filename, ext, data):
    write_dir = "/home/rx_pi/Downloads/"+filename+ext    
    with open(write_dir, "wb") as file:
        print(write_dir)
        print("#########################################")
        file.write(bytes(data, 'latin1'))

def binary2File_2(data):
    ext = data[0:16]
    data = data[16:]
    ext = ext.rstrip('\x00')
    filename = data.split(':ÿªÿ:')[0]
    data = data.split(':ÿªÿ:')[1]
    print("#########################################\nFile saved in :")
    writeFile_2(filename,ext, data)
```

```python
import time
t = 20000000 #5ms

prev_ip = count0 = count1 = count = oldcount1 = oldendtime = 0
oldcount0 = 11

data = bindata = rcv_data = marker_seq = bit = ''

marker_detect_start = communication_start = communication_end = start = False
communication_end = False

starttime = time.perf_counter_ns()
while True:
    ip = GPIO.input(inpin)
    if(prev_ip != ip):
        if ( (count0 <5) &(count1 < 5) ):
            count0 = oldcount0
            count1 = oldcount1
            endtime = oldendtime
            print('billa')
        else:
            endtime = time.perf_counter_ns()
            difftime = endtime - starttime
            bit = round(difftime/t) * str(prev_ip)
            if (bit != ''):
                count += 1
                #print(bit,count)
            oldcount0 = count0
            oldcount1 = count1
            oldendtime = endtime
            count0 = count1 = 0
            prev_ip = ip
            starttime = endtime
            bindata += bit
            
            
    if(ip == 0):
        count0 += 1
    else:
        count1 += 1
    endtimebtn = time.perf_counter_ns()
    if (endtimebtn - starttime)>t:
        starttime = endtimebtn
        count += 1
        bit = str(prev_ip)
        bindata += str(prev_ip)
        #print(prev_ip,count)
    
     # ----- THIS 'IF' BLOCK CHECKS THE START OF COMMUNICATION -----
    if ( (bit == '1') & (not start) ):
        start = True
        bindata = '1'
        count = 1
        i = 1
        starting_timeRx = time.perf_counter()
        
    # ----- THIS 'IF' BLOCK GATHERS '8bit' TO CONVERT THEM TO ASCII EQUIVALENT -----
    if (count == 8):
        letter = bin2Data(bindata)
        if (letter == "ÿ"):                                 # DETECTS STARTING LETTER OF MARKER 'ÿ'
            marker_detect_start = True
        if (marker_detect_start & (marker_seq in "ÿ/|\ÿ")):
            marker_seq += letter
            if ((marker_seq == "ÿ/|\ÿ")):                   # ONCE FULL MARKER IS DETECTED, IT WILL EITHER START OR END -
                print('Marker Received')                                                                #- COMMUNICATION
                if (not communication_start):
                    communication_start = True
                else:
                    communication_start = False
                    communication_end = True
                marker_detect_start = False
                marker_seq = letter = ''
        else:
            marker_detect_start = False
            marker_seq = ''
        if (communication_start & (not communication_end)): # ONCE COMMUNICATION STARTED, IT WILL ACCUMULATE LETTER BY LETTER
            rcv_data += letter
        if communication_end:                               # ONCE COMMUNICATION ENDS, IT SAVES DATA
            rcv_data = rcv_data.rstrip('ÿ/|\\')
            data01 = rcv_data
            binary2File_2(rcv_data)
            rcv_data=''
            ending_timeRx = time.perf_counter()
            print("Time taken for the image transfer : "+str(round(ending_timeRx - starting_timeRx))+"seconds")
            communication_end = False
            start = False
        print(bindata, letter, marker_detect_start, communication_start, marker_seq)
        bindata = ''
        count = 0
```
